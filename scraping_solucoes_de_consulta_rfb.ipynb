```python
import requests
from bs4 import BeautifulSoup
import pdfplumber
import re

# Lista de URLs para scraping
urls = [
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=97860",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=116317",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=133463",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=134370",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=133214",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=128418",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=137232",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=132739",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=140577",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=129674",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?naoPublicado=&idAto=139308&visao=anotado",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=130068",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=133752",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=132106",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=132651",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=130011",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=118766",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=139439",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=131615",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=118763",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=132742",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=89170",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=138976",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=84694",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=136846",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=118606",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=136938",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=53105",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=115875",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=120710",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=110736",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=133751",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=133746",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=116526",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=97808&visao=anotado",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=60118",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=77088#:~:text=SC%20Cosit%20n%C2%BA%2015%2F2016&text=Conclui%2Dse%20que%20o%20consulente,que%20se%20refere%20o%20art.",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=89166",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=96799",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=112720",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=97797",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=99659",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=71818",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=132738",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=132900",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=99742",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=63309&visao=anotado",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=95342&visao=anotado#:~:text=SC%20Cosit%20n%C2%BA%20164%2F2018&text=EMENTA%20Despesas%20com%20locomo%C3%A7%C3%A3o%20e,o%20%C3%B4nus%20tenha%20sido%20deste.",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=95202",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=64364",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=51833",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=116492",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=132871",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=83295",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=84136",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=56599",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=110728",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=58814",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=99665",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=114495",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=133216",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=124108",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=115845",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=89092",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=102043",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=53346&visao=anotado",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=130197",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=120541",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=54516",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=89111",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=122399",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=125158",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=120776",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=122249",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=88890&visao=anotado",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=120714",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=120711",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=122414",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=48868",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=120740",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=51491",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=79704",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=111543",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=98003",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=59829",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=114755",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=49683",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=58807",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=52786",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?visao=anotado&idAto=95378",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=98794",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=101697",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=133537",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=101911",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=82439",
    "http://normas.receita.fazenda.gov.br/sijut2consulta/link.action?idAto=112717&visao=anotado"
]

# Função para extrair o título da página
def get_page_title(soup):
    title_element = soup.find('title')
    if title_element:
        return title_element.text.strip()
    return "Título não encontrado"

# Função para processar cada link
def process_link(url):
    try:
        # Fazer a requisição para a página
        response = requests.get(url)
        response.raise_for_status()

        # Carregar o conteúdo da página no BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extrair o link do PDF
        pdf_link = soup.find('a', href=re.compile("anexoOutros.action"))
        if pdf_link:
            pdf_url = "http://normas.receita.fazenda.gov.br/sijut2consulta/" + pdf_link['href']
        else:
            print(f"Link do PDF não encontrado na página: {url}")
            return

        # Baixar o PDF
        pdf_response = requests.get(pdf_url)
        pdf_response.raise_for_status()

        # Extrair o título da página
        page_title = get_page_title(soup)

        # Limpeza do título para uso como nome de arquivo
        safe_title = re.sub(r'[^\w\s-]', '', page_title).replace(" ", "_")

        # Salvar o PDF localmente com o título como nome
        pdf_filename = f"{safe_title}.pdf"
        with open(pdf_filename, "wb") as pdf_file:
            pdf_file.write(pdf_response.content)

        # Converter o PDF para texto e salvar com o título no .txt
        txt_filename = f"{safe_title}.txt"
        with pdfplumber.open(pdf_filename) as pdf:
            with open(txt_filename, "w", encoding="utf-8") as txt_file:
                # Escrever o título no início do arquivo
                txt_file.write(f"{page_title}\n\n")

                # Percorrer as páginas e extrair o texto
                for page in pdf.pages:
                    text = page.extract_text()
                    if text:
                        txt_file.write(text)
                        txt_file.write("\n\n")  # Espaço entre páginas

        print(f"O arquivo '{txt_filename}' foi gerado com sucesso.")
    except Exception as e:
        print(f"Erro ao processar o link {url}: {e}")

# Iterar sobre cada URL na lista
for url in urls:
    process_link(url)
```
