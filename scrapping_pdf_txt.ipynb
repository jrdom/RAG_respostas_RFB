```python
import pandas as pd
import pdfkit
import fitz  # PyMuPDF
import os

path_wkhtmltopdf = r'D:\wkhtmltopdf\bin\wkhtmltopdf.exe'

config = pdfkit.configuration(wkhtmltopdf=path_wkhtmltopdf)

df = pd.read_excel("frequencia_normas_padronizadas - link.xlsx")
```

```python
df.head()
```

```python
for index, row  in df.iterrows():
    url = str(row["Link:"])
    pdfkit.from_url(url, f"normas_pdf/{str(row['Norma']).replace('/', '')}.pdf", configuration=config)
```

```python
folder_path = 'normas_pdf'
files = os.listdir(folder_path)
```

```python
for file in files:

    # Caminho do PDF
    pdf_path = os.path.join(folder_path, file)

    # Nome da pasta de saída
    output_folder = "normas_txt"
    os.makedirs(output_folder, exist_ok=True)  # Cria a pasta se não existir

    # Abrir o PDF
    pdf_document = fitz.open(pdf_path)

    # Extrair todo o texto do PDF
    texto_completo = ""
    for page_num in range(pdf_document.page_count):
        page = pdf_document[page_num]
        texto_completo += page.get_text()

    # Separar o texto em artigos
    artigos = [artigo.strip() for artigo in texto_completo.split("Art.") if artigo.strip()]
    artigos = [f"Art. {artigo}" for artigo in artigos]  # Recoloca "Art." no início de cada artigo

    # Salvar cada artigo em um arquivo TXT separado dentro da pasta "normas_txt"
    for i, artigo in enumerate(artigos, start=1):
        nome_arquivo = os.path.join(output_folder, f"{file.replace('.pdf','')}.txt")
        with open(nome_arquivo, "w", encoding="utf-8") as arquivo:
            arquivo.write(artigo)
        print(f"Artigo {i} salvo como {nome_arquivo}")

    # Fechar o PDF
    pdf_document.close()
```

```python
# Nome da pasta de saída
output_folder = "acordaos_txt"
os.makedirs(output_folder, exist_ok=True)  # Cria a pasta se não existir

# Itera sobre todos os arquivos na pasta
for file in os.listdir(folder_path):
    if file.endswith(".pdf"):
        # Caminho do PDF
        pdf_path = os.path.join(folder_path, file)

        # Abrir o PDF
        pdf_document = fitz.open(pdf_path)

        # Extrair todo o texto do PDF
        texto_completo = ""
        for page_num in range(pdf_document.page_count):
            page = pdf_document[page_num]
            texto_completo += page.get_text()

        # Nome do arquivo TXT
        nome_arquivo = os.path.join(output_folder, f"{file.replace('.pdf', '')}.txt")
        
        # Salvar o texto completo em um único arquivo TXT
        with open(nome_arquivo, "w", encoding="utf-8") as arquivo:
            arquivo.write(texto_completo)
        print(f"Texto completo salvo como {nome_arquivo}")

        # Fechar o PDF
        pdf_document.close()
```
